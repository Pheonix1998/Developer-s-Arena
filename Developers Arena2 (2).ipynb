```python
# Simple Student Grade Calculator

# Ask the user to enter their marks
marks = float(input("Enter your marks (out of 100): "))

# Check the grade based on the marks entered
if marks >= 90:
    grade = "A"
    message = "Excellent work! Keep it up! ðŸŒŸ"
elif marks >= 75:
    grade = "B"
    message = "Great job! You're doing really well! ðŸ‘"
elif marks >= 60:
    grade = "C"
    message = "Good effort! You can do even better! ðŸ’ª"
elif marks >= 45:
    grade = "D"
    message = "You passed! Try to aim higher next time! ðŸ™‚"
else:
    grade = "F"
    message = "Don't give up! Study hard and you'll improve! ðŸ’«"

# Display the final grade and message
print("\nYour Grade:", grade)
print(message)
```
